```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set()
```

```python
data =pd.read_csv('/content/2. Multiple linear regression 2.csv')
```

```python
data.head()
```

```python
data.describe()
```

```python
y=data['GPA']
x1= data[['SAT','Rand 1,2,3']]
```

```python
x=sm.add_constant(x1)
```

```python
results=sm.OLS(y,x).fit()
results.summary()
```
